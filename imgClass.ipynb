                                                        LOADING THE DEEPLEARNING MODEL

```python
mobile = tf.keras.applications.mobilenet_v2.MobileNetV2()
```

                                                        STATICALLY INPUTING THE IMAGE PATH

```python
filename = 'images//deerImg.jpg'
```

                                                            SHOWING THE IMAGE

```python
from tensorflow.keras.preprocessing import image
img = image.load_img(filename,target_size = (224,224))
```

```python
plt.imshow(img)
```

                                                            PRE-PROCESSING THE IMAGE

```python
resized_img = image.img_to_array(img)
print(resized_img.shape)
final_img = np.expand_dims(resized_img,axis = 0)
print(final_img.shape)
final_img = tf.keras.applications.mobilenet_v2.preprocess_input(final_img)
```

```python
predictions = mobile.predict(final_img)
```

```python
from tensorflow.keras.applications import imagenet_utils   #IMPORTING THE IMAGENET_UTILS
```

```python
results = imagenet_utils.decode_predictions(predictions) #ASSIGNING DECODED PREDICTIONS IN TO RESULTS
```

                                                            PRINTING THE RESULTS

```python
print(results) ## printing the results
```

                                                            DETECTING THE IMAGE

```python
print(results[0][0][1]) #printing the final result
```
